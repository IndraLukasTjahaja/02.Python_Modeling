```python
## Missing Values

# remove rows with nan
temp = df.dropna()

test = df['Gender'].dropna()

print(test)
print(temp.info())
```

```python
# fill na with 0
temp = df.fillna(0)
print(temp.info())

print(temp.groupby(['Gender']).count())
```

```python
df_missing = df
#fill the missing values with select data.
df_missing["Gender"] = df_missing["Gender"].fillna("b")
# replace the missing value with median or mean
df_missing["Age"].fillna(df_missing["Age"].median(), inplace=True)
# replace the others missing values variabel
df_missing["Marital_Status"] = df_missing["Marital_Status"].fillna("u")
df_missing["Bank_Customer"] = df_missing["Bank_Customer"].fillna("g")
df_missing["Education_Level"] = df_missing["Education_Level"].fillna("c")
print(df_missing.info())
```

```python
## Transformation - Change String to Numbers

df_transform = df_missing
# Create dummy variables for String type of data
temp  = pd.get_dummies(df_transform['Gender'], prefix='Gender')
print(temp)
```

```python
#Untuk dummy variable, kita tidak perlu dua faktor Gender_a dan Gender_b, pilih salah satu
temp.drop(['Gender_a'], axis=1, inplace=True)
print(temp.info())
```

```python
#combine the dummy variable back to the main dataframe and remove the old factor: gender
df_transform = df_transform.join(temp) #join the dummies to main table
df_transform = df_transform.drop(['Gender'], axis=1) # remove the String variabel
print(df_transform.info())
```

```python
# Dummy Variables - untuk sisa dari faktor yang string
temp  = pd.get_dummies(df_transform['Driver_License'], prefix='Driver_License')
temp.drop(['Driver_License_f'], axis=1, inplace=True)
df_transform = df_transform.join(temp) #join the dummies to main table
df_transform = df_transform.drop(['Driver_License'], axis=1) # remove the String variabel
temp  = pd.get_dummies(df_transform['Citizen'], prefix='Citizen')
temp.drop(['Citizen_p'], axis=1, inplace=True)
df_transform = df_transform.join(temp) #join the dummies to main table
df_transform = df_transform.drop(['Citizen'], axis=1) # remove the String variabel
temp  = pd.get_dummies(df_transform['Bank_Customer'], prefix='Bank_Customer')
temp.drop(['Bank_Customer_gg'], axis=1, inplace=True)
df_transform = df_transform.join(temp) #join the dummies to main table
df_transform = df_transform.drop(['Bank_Customer'], axis=1) # remove the String variabel
temp  = pd.get_dummies(df_transform['Marital_Status'], prefix='Marital_Status')
temp.drop(['Marital_Status_l'], axis=1, inplace=True)
df_transform = df_transform.join(temp) #join the dummies to main table
df_transform = df_transform.drop(['Marital_Status'], axis=1) # remove the String variabel
```

```python
## Transformation - Classify several strings to one group

# Make group for various Education Level
Education_Level_Category =    {
                        'cc': 'high',
                        'q': 'high',
                        'r': 'high',
                        'x': 'high',
                        'aa': 'low',
                        'd': 'low',
                        'ff': 'low',
                        'i': 'low',
                        'j': 'low',
                        'k': 'low',
                        'c': 'normal',
                        'e': 'normal',
                        'm': 'normal',
                        'w': 'normal',
                    }
df_transform['Education_Level_Group'] = df_transform['Education_Level'].map(Education_Level_Category)
print(df_transform.head())
```

```python
#dari faktor grouping education level, harus kembali dibuat jadi dummy variable
temp  = pd.get_dummies(df_transform['Education_Level_Group'], prefix='Education_Level_Group')
temp.drop(['Education_Level_Group_normal'], axis=1, inplace=True)
df_transform = df_transform.join(temp) #join the dummies to main table
df_transform = df_transform.drop(['Education_Level_Group','Education_Level'], axis=1) # remove the String variabel
print(df_transform.head())
```

```python
## Outliers - Apakah data input salah?

# we can create our own function to change max or min
def change_max(baris):
    if (baris['Years_Employed'] >= 50):
        return 50
    elif (baris['Years_Employed'] < 50):
        return baris['Years_Employed']
    else:
        return -1

temp['contoh'] = df_transform.apply(change_max, axis=1)
print(temp)

# or use the following code
df_transform.loc[df_transform['Years_Employed'] > 50, 'Years_Employed'] = df_transform["Years_Employed"].median()
```

```python
## Variable Selection
# Remove variables that not used in analyst
df_transform = df_transform.drop(['Zip_Code'], axis=1) # remove the String variabel
```

```python
## Machine Learning Part

# machine learning library
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

#to check performance of model
from sklearn.metrics import accuracy_score

#plot
import matplotlib.pyplot as plt
```

```python
# Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_Train, Y_Train)
logreg.score(X_Train, Y_Train)
Y_pred_log = logreg.predict(X_Test)

# Checking the model's accuracy
accuracy_score(Y_Test,logreg.predict(X_Test))
```
